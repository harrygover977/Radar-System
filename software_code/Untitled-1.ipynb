```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 

# magic command to display plots inline 
%matplotlib inline 

# set sty;e for better looking plots
plt.style.use('default')
```

__Key Concepts:__
- __Figure__: The top level container (the entire window/page)
- __Axes__: The actual plot (area where data is draw)
- __Axis__: The number line objects (x-axis, y-axis)

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

for i in range(len(x)):
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.title("Simple Line Plot")
    plt.scatter(x,y)
    plt.pause(0.1)
# not needed in notebook but is in standard python file 
plt.show()
```

```python
# Method 2: Object-oriented approach (more control recomended for complex plots)
fig, ax = plt.subplots() # creates a figure and axes object 
ax.plot(x,y)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Line Plot (Object-Oriented)')
plt.show()
```

```python
x = np.linspace(0, 10, 100) # 100 points from 0 to 10
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(2*x)

plt.figure(figsize=(8, 5)) # figsize=(wi9dth, height) in inches
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='cos(2x)')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title("Sine Wave")
plt.legend()
plt.grid(True, alpha=0.5) # Add a grid for easier reading
plt.show()
```

__Formatting__

__common format string__: [colour][marker][line]
- __Colours__: b (blue), r (red), g (green), c (cyan), m (magenta), y (yellow), k (black)
- __Markers__: o (circle), s (square), ^ (triangle), * (star), + (plus), x (x)
- __Lines__: - (solid), -- (dashed), -. (dash-dot), : (dotted)

```python
df = pd.read_csv('data.csv')

plt.plot(df['Position'], df['Distance'])
plt.xlabel("Position")
plt.ylabel("Distance")
plt.title("Sensor readings")
plt.show()
```
